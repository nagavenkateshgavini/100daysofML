# Problem Statement

In this exercise you'll try to build a neural network that predicts the salary of an employee based on their number of years of experience.

So, imagine if estimating an employees was as extremely simple. Base pay of 50k for someone with no experience and every additional year of experience gives you an additional 10k.

How would you create a neural network that learns this relationship so that it would predict that an employee with 8 years of experience should earn 130k.

# Solution

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
def prepare_data():
    xs = [0, 1, 2, 3, 4, 5, 6] # years if experience
    ys = [0.5, 0.6, 0.7, 0.8, 0.9, 1, 1.1] # salary
    
    return [xs, ys]
```

```python
def get_model():
    xs, ys = prepare_data()
    model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    model.compile(optimizer="sgd", loss="mean_squared_error")
    model.fit(xs, ys, epochs=100)
    
    return model
```

```python
def predict_salary(years_of_experience):
    model = get_model()
    pred = model.predict(years_of_experience)[0][0]
    return pred
```

```python
exp = [8.0]
salary_pred = predict_salary(exp)
print(f"For {exp} of experience, you will get {salary_pred * 100} Thousand Dollars")
```
